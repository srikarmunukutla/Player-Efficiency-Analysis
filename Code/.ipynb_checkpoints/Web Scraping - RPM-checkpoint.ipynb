```python
# from selenium import webdriver
# from selenium.webdriver.common.keys import Keys
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://www.espn.com/nba/statistics/rpm/_/year/2020"
page = requests.get(url)
soup = BeautifulSoup(page.text, "html.parser")

header = soup.find("tr", attrs = {"class": "colhead"})
columns = [col.get_text() for col in header.find_all("td")]
columns = [columns[1]] + columns[5:]
columns[-1] = "RPM WINS"

df_rpmstats = pd.DataFrame(columns=columns)
df_rpmstats
```

```python
columns
```

```python
players = soup.find("tr", attrs = {"class":re.compile("row")})
players
for player in players.find_all("td"):
    print(player.get_text())
```

```python
total_results = soup.find("div", attrs = {"class": "totalResults"})
num_players = int(total_results.get_text().split(" ")[0])
interval = 40

for a in range(1, num_players, interval):
    page_url = url + "/page/{}".format(int(a / interval)+1)
    page = requests.get(page_url)
    soup = BeautifulSoup(page.text, "html.parser")

    players = soup.find_all("tr", attrs = {"class": re.compile("row")})
    for player in players:
        rpmstats = []
        i = 0
        for data in player.find_all("td"):
            string = data.get_text()
            if i == 1:
                name = string.split(",")[0]
                rpmstats.append(name)
            elif i >= 5:
                rpmstats.append(string)
            i += 1

        temp_df = pd.DataFrame(rpmstats).transpose()
        print(temp_df)
        temp_df.columns = columns

        df_rpmstats = pd.concat([df_rpmstats, temp_df], ignore_index=True)

df_rpmstats
```

```python
df_rpmstats.to_csv("../Data/rpm18-19.csv", index=False, sep=',', encoding='utf-8')
```

```python
df_rpmstats = pd.read_csv("../Data/rpm18-19.csv")
final_df = pd.read_csv("../Data/alldata18-19.csv")
```

```python
for i in range(1, len(columns)):
    final_df.insert(len(final_df.columns), columns[i], int(0))
# number of columns is also len(final_df.columns)
```

```python
for ind in range(final_df.shape[0]):
    for row in range(df_rpmstats.shape[0]):
        if final_df.loc[ind, "NAME"] == df_rpmstats.loc[row, "NAME"]:
            final_df.loc[ind, "ORPM"] = df_rpmstats.loc[row, "ORPM"]
            final_df.loc[ind, "DRPM"] = df_rpmstats.loc[row, "DRPM"]
            final_df.loc[ind, "RPM"] = df_rpmstats.loc[row, "RPM"]
            final_df.loc[ind, "RPM WINS"] = df_rpmstats.loc[row, "RPM WINS"]
            break
final_df.head()
```
