```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("../Data/alldata18-19.csv")
```

```python
data.head()
```

```python
salstats = pd.DataFrame()
salstats.insert(0, "NAME", data.iloc[:, 0])
salstats.head()
```

```python
temp = (data.loc[:, "SALARY"]/data.loc[:, "Wins Above Replacement"].replace(to_replace=0.0, value=0.0001))
for row in temp:
    print(row)

# print("===========")
# temp.replace(to_replace=float('inf'), value=0.1, inplace=True)
# for row in temp:
#     print(row)
```

```python
salstats.insert(salstats.shape[1], "Salary by PER", (data.loc[:, "SALARY"]/data.loc[:, "PER"].replace(to_replace=0.0, value=0.0001)))
salstats.insert(salstats.shape[1], "Salary by WAR", (data.loc[:, "SALARY"]/data.loc[:, "Wins Above Replacement"].replace(to_replace=0.0, value=0.0001)))
salstats.insert(salstats.shape[1], "Salary by WS", (data.loc[:, "SALARY"]/data.loc[:, "Win Shares"].replace(to_replace=0.0, value=0.0001)))
salstats.insert(salstats.shape[1], "Salary by BPM", (data.loc[:, "SALARY"]/data.loc[:, "BPM"].replace(to_replace=0.0, value=0.0001)))
salstats.insert(salstats.shape[1], "Salary by RPM", (data.loc[:, "SALARY"]/data.loc[:, "RPM"].replace(to_replace=0.0, value=0.0001)))
```
