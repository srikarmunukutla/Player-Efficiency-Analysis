```python
# from selenium import webdriver
# from selenium.webdriver.common.keys import Keys
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.basketball-reference.com/leagues/NBA_2019_advanced.html"
page = requests.get(url)
soup = BeautifulSoup(page.text, "html.parser")

columns = ["NAME", "OWS", "DWS", "Win Shares", "WS/48", "OBPM", "DBPM", "BPM", "VORP", "Wins Above Replacement"]

df_winshares = pd.DataFrame(columns=columns)
df_winshares
```

```python
players = soup.find("tr", attrs = {"class": "full_table"})
# players
i = 0
for player in players.find_all("td"):
#     i += 1
#     if i == 24:
#         continue
#     if i >= 20:
    print(player.get_text())
```

```python
players = soup.find_all("tr", attrs = {"class": "full_table"})
for player in players:
    wsstats = []
    i = 0
    for data in player.find_all("td"):
        i += 1
        if i == 1:
            wsstats.append(data.get_text())
        if i == 24:
            continue
        if i >= 20:
            wsstats.append(data.get_text())
    wsstats.append(float(wsstats[-1]) * 2.70)
    
    temp_df = pd.DataFrame(wsstats).transpose()
    temp_df.columns = columns

    df_winshares = pd.concat([df_winshares, temp_df], ignore_index=True)

df_winshares
```

```python
df_winshares.to_csv(r"/Users/Srikar/Desktop/Sports/Data/winshares18-19.csv", index=False, sep=',', encoding='utf-8')
```
